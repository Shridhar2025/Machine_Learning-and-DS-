```python
knn = KNeighborsClassifier(n_neighbors=12)

knn.fit(x3_train,y3_train)
pred_3 = knn.predict(x3_test)

print('WITH K=12')
print('\n')
print(confusion_matrix(y3_test,pred_3))
print('\n')
print(classification_report(y3_test,pred_3))
```
